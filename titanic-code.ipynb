***Importing necesseary libraries***

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

*Loading our data*

```python
df_train=pd.read_csv("/kaggle/input/titanic-dataset/train.csv")
df_test=pd.read_csv("/kaggle/input/titanic-dataset/test.csv")
df_test.head()
```

**Data preprocessing**

```python

df_train["Age"].fillna(df_train["Age"].mean(), inplace=True)

df_train["Embarked"].dropna()

df_test["Age"].fillna(df_test["Age"].mean(), inplace=True)

df_test["Fare"].fillna(df_test["Fare"].mean(),inplace=True)
print(df_train.isnull().sum())
```

**Enconding the necessary columns**

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the encoder
label_encoder = LabelEncoder()
#encoding the necessary colmumns for the training dataframe

df_train['Cabin'] = label_encoder.fit_transform(df_train['Cabin'])
df_train['Name']=label_encoder.fit_transform(df_train['Name'])
df_train['Sex']=label_encoder.fit_transform(df_train['Sex'])
df_train["Ticket"]=label_encoder.fit_transform(df_train['Ticket'])
df_train['Embarked']=label_encoder.fit_transform(df_train['Embarked'])
#encoding the necessary columns for the testing dataframe
df_test['Cabin'] = label_encoder.fit_transform(df_test['Cabin'])
df_test['Name']=label_encoder.fit_transform(df_test['Name'])
df_test['Sex']=label_encoder.fit_transform(df_test['Sex'])
df_test["Ticket"]=label_encoder.fit_transform(df_test['Ticket'])
df_test['Embarked']=label_encoder.fit_transform(df_test['Embarked'])
```

***Splitting the data***

```python
X_train=df_train.drop("Survived",axis=1)
y_train=df_train["Survived"]
```

****Training the Model****

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
# Make predictions
y_pred = model.predict(df_test)
output = pd.DataFrame({'PassengerId': df_test['PassengerId'], 'Survived': y_pred})
output.to_csv('submission.csv', index=False)
```
